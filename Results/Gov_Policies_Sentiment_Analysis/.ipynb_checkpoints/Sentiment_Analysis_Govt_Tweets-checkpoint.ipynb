# **Extracting dataframe**

**Used Header names**

C = **TweetId**, D = **TweetCreatedAt**, E = **TweetText**,        G = **TweetUserMentions**, Q = **isWHO**, R = **isCDCgov**, S = **isDonaldTrump**, T = **isCDCEmergency**, 
U = **isNIAIDNews**, V = **IsV2019N**, W = **date**, **WC**, **affect**, **posemo**, **negemo**, **anx**, **anger**, **sad**

```python
df = pd.read_csv("GovPoliciesTweetsWithDate_liwc.csv", usecols=["C", "D", "E", "G", "Q", "R",	"S", "T", "U", "V", "W", "WC", "affect", "posemo", "negemo", "anx", "anger", "sad"])

df = df[1:] #take the data less the header row
df.dropna(inplace=True)
df = df.drop_duplicates(subset='C')
df = df.set_index('C')
df
```

```python
def calculate_polarity(x):
    blob = TextBlob(x)
    return blob.sentiment.polarity
```

```python
df['PolarityScore'] = df['E'].apply(calculate_polarity)
df
```

```python
def calculate_subjectivity(x):
    blob = TextBlob(x)
    return blob.sentiment.subjectivity
```

```python
df['SubjectivityScore'] = df['E'].apply(calculate_subjectivity)
df
```

```python
def get_english_tweets(tweetContent):
  return detect(tweetContent) == "en"

#df = df['E'].apply(get_english_tweets)

df['Lang'] = df['E'].apply(get_english_tweets)
df
```

```python
df = df[df.eval("Lang == True")]
df
```

# **Sentiment plots: Average score v/s Government twitter ids**

```python
govt_mentions = ['WHO', 'CDCgov', 'DonaldTrump', 'CDCEmergency', 'NIAIDNews', 'V2019N']
```

```python
pos_sentiments = []
#WHO_pos = govt_WHO_df['posemo'].mean()
pos_sentiments.append(govt_WHO_df['posemo'].mean())
pos_sentiments.append(govt_CDCgov_df['posemo'].mean())
pos_sentiments.append(govt_DonaldTrump_df['posemo'].mean())
pos_sentiments.append(govt_CDCEmergency_df['posemo'].mean())
pos_sentiments.append(govt_NIAIDNews_df['posemo'].mean())
pos_sentiments.append(govt_V2019N_df['posemo'].mean())
print(pos_sentiments)
```

```python
neg_sentiments = []
neg_sentiments.append(govt_WHO_df['negemo'].mean() * -1)
neg_sentiments.append(govt_CDCgov_df['negemo'].mean() * -1)
neg_sentiments.append(govt_DonaldTrump_df['negemo'].mean() * -1)
neg_sentiments.append(govt_CDCEmergency_df['negemo'].mean() * -1)
neg_sentiments.append(govt_NIAIDNews_df['negemo'].mean() * -1)
neg_sentiments.append(govt_V2019N_df['negemo'].mean() * -1)
print(neg_sentiments)
```

**Positive and negative sentiment scores plot**

```python
plt.bar(govt_mentions, neg_sentiments)
plt.bar(govt_mentions, pos_sentiments)
plt.xticks(rotation = 90)
plt.xlabel('Government twitter ids')
plt.ylabel('Negative Positive sentiment score')
plt.figure(figsize=(10,20))
plt.show()
```
