### Data preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

**** Cleaned data defining number of confirmed cases for every country from 22nd Jan to 31st March

```python
def cleaningGlobalConfirmedData():
    df = pd.read_csv("time_series_covid19_confirmed_global.csv")
    df.drop(["Province/State", "Lat", "Long"], axis = 1, inplace = True)
    df_confirmed = df.groupby(['Country/Region'], as_index=False).sum()
    df_confirmed.to_csv(r'global_confirmed.csv', index=True)
    return df_confirmed
```

```python
#Looping over all the unique entries in the grouped dataframe
def plottingConfirmedCasesData(dataframe):
    for test in range(0,178):
        #Getting each row and storing in a new dataframe
        abra = dataframe.loc[test]
        #Getting the country name from each row
        country = abra.iloc[0]
        #Splicing the dataframe to get the y values - total no of confirmed cases across days
        x = abra.iloc[1::10]#Taking data for every 10 days
        #Setting the plot title
        title = "COVID-19 cases in "+country
        #Plotting each country's plot
        plt.plot(x)
        #Setting the plot title
        plt.title(title)
        #Setting the x-axis label for the plot
        plt.xlabel("Dates")
        #Setting the y-axis label for the plot
        plt.ylabel("No. of cases")
        #Display all the graphs
        plt.show()
```

**** Cleaned data defining number of recovered patients for every country from 22nd Jan to 31st March

```python
globalConfirmedCases = cleaningGlobalConfirmedData()
globalConfirmedCases
```

```python
plottingConfirmedCasesData(globalConfirmedCases)
```

```python
def cleaningGlobalRecoveredData():
    df = pd.read_csv('time_series_covid19_recovered_global.csv')
    df.drop(["Province/State", "Lat", "Long"], axis = 1, inplace = True)
    recovered_data = df.groupby(['Country/Region'], as_index=False).sum()
    recovered_data.to_csv(r'global_recovered.csv', index = True)
    return recovered_data
```

**** Cleaned data defining number of deaths for every country from 22nd Jan to 31st March

```python
globalRecoveredCases = cleaningGlobalRecoveredData()
globalRecoveredCases
```

```python
# Cleaned data defining number of deaths for every country from 22nd Jan to 31st March
def cleaningGlobalDeathsData():
    globalDeathsCSV = pd.read_csv("time_series_covid19_deaths_global.csv")
    globalDeathsCSV.drop(['Province/State', 'Lat', 'Long'], axis=1, inplace = True)
    finalDeathsData = globalDeathsCSV.groupby(['Country/Region'], as_index=False).sum()
    finalDeathsData.to_csv (r'global_deaths.csv', index = True)
    return finalDeathsData
```

```python
globalDeathsCases = cleaningGlobalDeathsData()
globalDeathsCases
```
