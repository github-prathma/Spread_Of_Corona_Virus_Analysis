```python
import pandas as pd

def cleaningGlobalDeathsData():
    globalDeathsCSV = pd.read_csv("time_series_covid19_deaths_global.csv")
    removeColumns = globalDeathsCSV.drop(['Province/State', 'Lat', 'Long'], axis=1)
    allCountriesFinalData = removeColumns.groupby(['Country/Region'], as_index=False).agg('sum')
    allCountriesFinalData.to_csv (r'global_deaths.csv', index = True, header=True)
```

```python
def globalRecoveredData():
  df = pd.read_csv('time_series_covid19_recovered_global.csv')
  rf = df.drop(['Province/State','Lat','Long'], axis = 1)
  recovered_data = rf.groupby(['Country/Region']).sum()
  recovered_data.to_csv(r'cleaned_global_recovered.csv', index = True)
```

```python
#Using pandas & numpy library for performing data analytics operations
import pandas as pd
import numpy as np
#Using matplotlib 
import matplotlib.pyplot as plt
def cleaningGlobalConfirmedData():
  #Reading data from the csv file
  df_confirmed = pd.read_csv("time_series_covid19_confirmed_global.csv")
  #Displaying first five entries of the dataframe
  df_confirmed.head()
  #Removing the unnecessary columns from the dataframe
  df_confirmed.drop(["Province/State", "Lat", "Long"], axis = 1, inplace = True)
  #Displaying the first five elements of dataframe to confirm the removal
  df_confirmed.head()
  #Grouping the contents of the dataframe based on the 'Country/Region' and adding the number of cases
  new_df = df_confirmed.groupby(['Country/Region'], as_index=False).sum()
  #Displaying the head of dataframe for verification purposes
  new_df.head()
  #Saving the cleaned data to a new csv file
  new_df.to_csv(r'cleaned_time_series_covid19_confirmed_global.csv', index=False)
  #Looping over all the unique entries in the grouped dataframe
  for test in range(0,178):
    #Getting each row and storing in a new dataframe
    abra = new_df.loc[test]
    #Getting the country name from each row
    country = abra.iloc[0]
    #Splicing the dataframe to get the y values - total no of confirmed cases across days
    x = abra.iloc[1::10]#Taking data for every 10 days
    #Setting the plot title
    title = "COVID-19 cases in "+country
    #Plotting each country's plot
    plt.plot(x)
    #Setting the plot title
    plt.title(title)
    #Setting the x-axis label for the plot
    plt.xlabel("Dates")
    #Setting the y-axis label for the plot
    plt.ylabel("No. of cases")
    #Display all the graphs
    plt.show()
```
